# 缺失值和唯一值

```python
print('There are {0} columns in train dataset with missing values.'.format(data.isnull().any().sum()))
```

```python
have_null_fea_dict = (data.isnull().sum() / len(data)).to_dict()
fea_null_moreThanHalf = {}
for key, value in have_null_fea_dict.items():
    if value > 0.5:
        fea_null_moreThanHalf[key] = value
fea_null_moreThanHalf
```

```python
missing = data.isnull().sum() / len(data)
missing = missing[missing > 0]
missing.sort_values(inplace=True)
missing.plot.bar()
```

# 查看数据类型

```python
numerical_fea = list(data.select_dtypes(exclude=['object']).columns)
category_fea = list(filter(lambda x: x not in numerical_fea, list(data.columns)))
```

```python
numerical_fea
```

```python
data.Timestamp
```

```python
data.Label
```

```python
def get_numerical_serial_fea(data, feas):
    numerical_serial_fea = []
    numerical_noserial_fea = []
    for fea in feas:
        temp = data[fea].nunique()
        if temp <= 10:
            numerical_noserial_fea.append(fea)
        else:
            numerical_serial_fea.append(fea)
    return numerical_serial_fea, numerical_noserial_fea
numerical_serial_fea, numerical_noserial_fea = get_numerical_serial_fea(data, numerical_fea)
```

```python
numerical_serial_fea
```

```python
numerical_noserial_fea
```

```python
for fea in numerical_noserial_fea:
    print(data[fea].value_counts())
```

```python
for fea in category_fea:
    print(data[fea].value_counts())
```

```python
plt.figure(figsize=(8, 8))
sns.barplot(data['Protocol'].value_counts(dropna=False).keys(),
            data['Protocol'].value_counts(dropna=False))
plt.show()
```

```python
f = pd.melt(data, value_vars=[fea for fea in numerical_serial_fea if fea not in ['Flow Byts/s', 'Flow Pkts/s']])
g = sns.FacetGrid(f, col="variable", col_wrap=4, sharex=False, sharey=False)
g = g.map(sns.distplot, "value")
```

# 数据间相关关系

```python
train_benign = data.loc[data['Label'] == 'Benign']
train_ddos = data.loc[data['Label'] != 'Benign']
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 8))
train_benign.groupby('Protocol')['Protocol'].count().plot(kind='barh', ax=ax1, title='Count of Protocol Benign')
train_ddos.groupby('Protocol')['Protocol'].count().plot(kind='barh', ax=ax2, title='Count of Protocol DDoS')
train_benign.groupby('FIN Flag Cnt')['FIN Flag Cnt'].count().plot(kind='barh', ax=ax3, title='Count of FIN Flag Cnt Benign')
train_ddos.groupby('FIN Flag Cnt')['FIN Flag Cnt'].count().plot(kind='barh', ax=ax4, title='Count of FIN Flag Cnt DDoS')
plt.show()
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 8))
train_benign.groupby('SYN Flag Cnt')['SYN Flag Cnt'].count().plot(kind='barh', ax=ax1, title='Count of SYN Flag Cnt Benign')
train_ddos.groupby('SYN Flag Cnt')['SYN Flag Cnt'].count().plot(kind='barh', ax=ax2, title='Count of SYN Flag Cnt DDoS')
train_benign.groupby('RST Flag Cnt')['RST Flag Cnt'].count().plot(kind='barh', ax=ax3, title='Count of RST Flag Cnt Benign')
train_ddos.groupby('RST Flag Cnt')['RST Flag Cnt'].count().plot(kind='barh', ax=ax4, title='Count of RST Flag Cnt DDoS')
plt.show()
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 8))
train_benign.groupby('PSH Flag Cnt')['PSH Flag Cnt'].count().plot(kind='barh', ax=ax1, title='Count of PSH Flag Cnt Benign')
train_ddos.groupby('PSH Flag Cnt')['PSH Flag Cnt'].count().plot(kind='barh', ax=ax2, title='Count of PSH Flag Cnt DDoS')
train_benign.groupby('ACK Flag Cnt')['ACK Flag Cnt'].count().plot(kind='barh', ax=ax3, title='Count of ACK Flag Cnt Benign')
train_ddos.groupby('ACK Flag Cnt')['ACK Flag Cnt'].count().plot(kind='barh', ax=ax4, title='Count of ACK Flag Cnt DDoS')
plt.show()
```
